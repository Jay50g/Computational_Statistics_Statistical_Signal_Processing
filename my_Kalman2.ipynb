```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
```

```python
A=np.array([0.99,0,0,0.99]).reshape(2,2)
```

```python
B=np.array([0.15,0.4])
B
```

```python
X=[]
X.append(np.zeros(2))
X[0]

```

```python
H=np.eye(2)
H
```

```python
Y=[]
Y.append(np.zeros(2))
Y[0]
```

```python
w=0.3*np.random.standard_normal(len(A))
```

```python
w
```

```python
W=np.diag(w)
```

```python
W
```

```python
v=0.3*np.random.standard_normal(len(A))
v
```

```python
V=np.diag(v)
V
```

```python
for i in range(1,100):
    W=0.3*np.random.standard_normal(len(A))
    V=0.3*np.random.standard_normal(len(A))
    X.append(A@X[i-1]+B+W)
    Y.append(X[i]+V)
```

```python
X=np.array(X)
Y=np.array(Y)
```

```python
plt.plot(X[:,0],X[:,1],'r.-')
plt.plot(Y[:,0],Y[:,1],'gx')
```

```python
R=np.diag([0.1,0.1])
P
Q=np.diag([0.1,0.1])
Q
```

```python
Xhat=[]
```

```python
x=np.array([0,0])
Xhat.append(x)
```

```python
for i in range(1,100):
    x=A@x+B
    P=A@P@A.T+Q
    K=P@H.T@np.linalg.inv(H@P@H.T+R)
    e=Y[i]-H@x
    x=x+K@e
    P=(np.eye(len(A))-K@H)@P
    Xhat.append(x)
    
Xhat=np.array(Xhat)
```

```python
plt.plot(X[:,0],X[:,1],'r.-')
plt.plot(Xhat[:,0],Xhat[:,1],'b--')
#plt.plot(Y[:,0],Y[:,1],'g-')
```
