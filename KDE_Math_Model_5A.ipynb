```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
mu=0
std=1
```

```python
sams=sp.stats.norm.rvs(mu,std,100)
```

```python
x=np.linspace(-3,3,1000)
```

```python
pdf=sp.stats.norm.pdf(x,mu,std)
cdf=sp.stats.norm.cdf(x,mu,std)
```

```python
plt.plot(x,pdf,'r-')
plt.plot(x,cdf,'b-')
plt.plot(sams,np.zeros(len(sams)),'ko')
```

```python
kde=sp.stats.gaussian_kde(sams)
```

```python
plt.plot(x,kde(x),'g-',lw=3)  # reconstructed statistics from samples
plt.plot(x,pdf,'r-',lw=3)  # Original statistics ( True values)
plt.legend(['Estimated Distribution','True Distribution'])
plt.ylim([0,1])
plt.grid()
```
