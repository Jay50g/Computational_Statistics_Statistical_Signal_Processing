```python
x=np.linspace(-4,4,1000)  
```

```python
sam=sp.stats.norm.rvs(0,1,100)  # u, std, sample_numbers
```

```python
pdf=sp.stats.norm.pdf(x)
```

```python
plt.plot(x,pdf,'r-')
#plt.plot(sam,[0]*len(sam),'b.')
plt.stem(sam,[0.1]*len(sam),basefmt=' ')
```

```python
kde=sp.stats.gaussian_kde(sam)
```

```python
plt.plot(x,pdf,'r-')
plt.stem(sam,[0.1]*len(sam),basefmt=' ')
plt.plot(x,kde(x),'g--')
```

```python
x=np.linspace(-5,5,1000)
```

```python
kde2=sp.stats.gaussian_kde(X1)
```

```python
plt.plot(x,kde2(x),'r-')
plt.plot(X1,[0]*len(X1),'bo')
plt.grid()
#plt.xlim((0.5,2.5))
#plt.ylim((0.4,0.5))
```

```python
n=len(X1)
```

```python
kernel_std=1.06*X1.std()/(n**(1/n))
```

```python
my_kde=np.zeros(len(x))
for i in X1:
    kernel=sp.stats.norm.pdf(x,i,kernel_std)/n
    plt.plot(x,kernel,'m-')
    my_kde+=kernel
plt.plot(x,my_kde,'g--')
plt.plot(x,kde2(x),'r-')
plt.plot(X1,[0]*len(X1),'bo')
plt.grid()
```

```python
x=np.linspace(-4,4,1000)
sam=sp.stats.norm.rvs(0,1,100)
pdf=sp.stats.norm.pdf(x,0,1)
#kde=sp.stats.gaussian_kde(sam)(x)
kde=sp.stats.gaussian_kde(sam)
kde=kde(x)
n=len(sam)
```

```python
kernel_std=0.4*sam.std()/(n**(1/n))
```

```python
plt.plot(x,pdf,'r-')
plt.plot(x,kde,'b-')
plt.plot(sam,[0]*n,'ko')
plt.stem(sam,[0.1]*n)

my_kde=np.zeros(len(x))
for i in sam:
    kernel=sp.stats.norm.pdf(x,i,kernel_std)/n  # x, mu, std
    plt.plot(x,kernel,'k-')
    my_kde+=kernel
plt.plot(x,my_kde,'g--')
```
