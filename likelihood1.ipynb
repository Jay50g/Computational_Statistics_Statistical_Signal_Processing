```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-5,5,500)
xD=sp.stats.norm.pdf(x,0,0.8)
```

```python
y=0.2
```

```python
pyOnx=sp.stats.norm.pdf(y,np.cos(6.28*x),1)
```

```python
plt.plot(x,pyOnx)
plt.grid()
plt.ylim([0,1])
plt.plot(x,xD,'r-',lw=3)
plt.plot(x,xD*pyOnx/0.3,'g-',lw=3)
```
