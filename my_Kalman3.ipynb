```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
F=np.array([0.99,0,0,0.99]).reshape(2,2)
F
```

```python
B=np.array([0.1,0.2])
B
```

```python
W=0.3*sp.stats.norm(0,1).rvs([2,1])
```

```python
W
```

```python
V=0.3*sp.stats.norm(0,1).rvs([2,1])
V
```

```python
H=np.eye(2)
H
```

```python
X=np.zeros((100,2))

Y=np.zeros((100,2))
Y
```

```python
for i in range(1,100):
    X[i]=F@X[i-1]+B+0.3*sp.stats.norm(0,1).rvs(2)
    Y[i]=H@X[i]+sp.stats.norm(0,0.3).rvs(2)
```

```python
plt.plot(X[:,0],X[:,1],'r.-')
plt.plot(Y[:,0],Y[:,1],'gx')
```
