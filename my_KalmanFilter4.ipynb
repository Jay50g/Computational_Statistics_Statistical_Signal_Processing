```python
# this example shows a random walk
# with a direct observator, ig. Y=X. 
```

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
F=np.array([0.99,0,0,0.99]).reshape(2,2)
F
```

```python
B=np.array([0.1,0.2])
B
```

```python
H=np.eye(2)
H
```

```python
X=np.zeros((100,2))
X
```

```python
Y=np.zeros([100,2])
Y
```

```python
V=sp.stats.norm(0,0.3).rvs(2) #np.random.normal(u,sig,size)
```

```python
Y[0]=H@X[0]+V
```

```python
for i in range(1,100):
    X[i]=F@X[i-1]+B+sp.stats.norm(0,0.3).rvs(2)
    Y[i]=H@X[i]+sp.stats.norm(0,0.4).rvs(2)
```

```python
p1=plt.plot(X[:,0],X[:,1],'r.-')
p2=plt.plot(Y[:,0],Y[:,1],'gx')
```

```python
Q=np.array([0.09,0.09])
R=np.array([0.16,0.16])
Q=np.diag(Q)
R=np.diag(R)
```

```python
R,Q
```

```python
p=Q.copy()
```

```python
XHat=np.zeros((100,2))
XHat
```

```python
X.shape
```

```python
XHat.shape
```

```python
XHat[0]=Y[0].copy()
XHat[0]=np.array([10,10])
```

```python
XHat[0]
```

```python
X[0]
```

```python
x=XHat[0].copy()
```

```python
# Kalman Filter ============
```

```python
for i in range(1,100):
    x=F@x+B+sp.stats.norm(0,0.3).rvs(2)  # xHat1Sub
    p=F@p@F.T+Q                          # pX1Sub
    K=p@H@np.linalg.inv(H@p@H.T+R)       #
    e=Y[i]-H@x
    x=x+K@e                              # xHat1
    p=(np.eye(2)-K@H)@p                  # pX1
    XHat[i]=x                            # save xHat1
```

```python
plt.plot(Y[:,0],Y[:,1],'gx')
plt.plot(X[:,0],X[:,1],'r.-')
plt.plot(XHat[:,0],XHat[:,1],'m-')
```

```python
chi=0
for i in range(10,100):
    chi=chi+((X[i]-XHat[i])**2)/XHat[i]
    
```
