```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
mu=np.pi/2,
sigma=1/2
n_particles=100
```

```python
particles=sp.stats.norm.rvs(mu,sigma,n_particles)
np.std(particles)
```

```python
kde=sp.stats.gaussian_kde(particles)
```

```python
x=np.linspace(-2.5,6.5,1000)
true=sp.stats.norm.pdf(x,mu,sigma)
estimate=kde(x)

h=0.4*np.std(particles)/(n_particles**(1/n_particles))

est=np.zeros(len(x))
for i in particles:
    kernel=sp.stats.norm.pdf(x,i,h)/n_particles
    est+=kernel

h
```

```python
plt.plot(x,true,'r-')
plt.plot(x,estimate,'k--')
plt.plot(x,est,'m--')
plt.stem(particles, [0.1]*len(particles), basefmt=' ')
plt.xlim([0,np.pi])
```

```python
mus= np.array([-1, -2, 3, 2, 1.5, 3.5,0])
n=len(mus)
h=1.06*np.std(mus)/(n**(1/n))
```

```python
x=np.linspace(-6,6,1000)
est=np.zeros(len(x))
```

```python
for mu in mus:
    kernel=sp.stats.norm.pdf(x,mu,h)/n
    plt.plot(x,kernel,'r-')
    est+=kernel
plt.plot(mus,[0]*n,'bo',markersize=15)
plt.plot(x,est,'k-')

kde=sp.stats.gaussian_kde(mus)
plt.plot(x,kde(x),'m-')
```
